DIABETES PREDICTION

Importing modules of python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Imporing the dataset using pandas


```python
diabetes_dataset = pd.read_csv("diabetes_prediction_dataset.csv")

```

```python
# printing the first 5 rows of the dataset

diabetes_dataset.head()
```

Converting the gender into interger values

```python
diabetes_dataset.drop('smoking_history',axis = 1,inplace = True)
```

```python
diabetes_dataset.head()
```

```python
def convertgender(g):
  if g == 'Male':
    return 1
  else :
    return 0
diabetes_dataset['gender'] = diabetes_dataset['gender'].map(convertgender)
diabetes_dataset.head()
```

```python
#printing the shpae of the data set
diabetes_dataset.shape
```

```python
#printing the info of the data set
diabetes_dataset.info()
```

```python
#describing the dataset
diabetes_dataset.describe
```

checking wether the data has any NaN values

```python
diabetes_dataset.isnull().sum()
# The data set doesn't have any null values so we are clear to go ahead
```

```python
# grouping the data set by diabetes
diabetes_dataset['diabetes'].value_counts()
```

0 --> non diabetic

1 --> diabetic


as the 0 values  and the 1 values of the data are in large difference we need to convert the 0 to that range
of the 1

```python
non_diabetic = diabetes_dataset[diabetes_dataset.diabetes == 0 ]
diabetic =diabetes_dataset[diabetes_dataset.diabetes == 1]
```

```python
print(non_diabetic.shape)
print(diabetic.shape)
```

converting the non_diabetic of 91500 to 8500


```python
non_diabetic_sample = non_diabetic.sample(n=8500)
print(non_diabetic_sample.shape)
```

concatinating the two data sets

```python
new_dataset = pd.concat([non_diabetic_sample,diabetic],axis = 0)
new_dataset.head()
```

Separating the target and the features

```python
X = new_dataset.drop(columns = 'diabetes',axis = 1)
Y = new_dataset['diabetes']
```

Standarding

```python
# creating a instance of the imported StandardScalar
scalar = StandardScaler()
scalar.fit(X)
```

```python
std_data = scalar.transform(X)
```

```python
X = std_data
Y = new_dataset['diabetes']
```

```python
#printing the feature values
print(X)
```

```python
#printing the targert values
print(Y)
```

Dividing the data into the training data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training the data model

```python
classifier = svm.SVC(kernel= 'linear')
# training the svm classifier
classifier.fit(X_train,Y_train)
```

MOdel evalution

Accuracy score

```python
# accuracy score on the training data
X_training_predict = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_training_predict,Y_train)
```

```python
print('Accuracy score is ',training_data_accuracy * 100)
```

```python
# accuracy score on the testing data
X_test_predict = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict,Y_test)
```

```python
print("accuracy score of testing data is ",test_data_accuracy * 100)
```
