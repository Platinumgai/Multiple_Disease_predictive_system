```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
heart_dataset = pd.read_csv('heart.csv')
```

0 --> non heart problem

```python

X = heart_dataset.drop(columns = 'target',axis = 1)
Y = heart_dataset['target']
```

```python
# standardizing the data
scalar = StandardScaler()
scalar.fit(X)
std_data = scalar.transform(X)

```

```python
X = std_data
```

```python
print(X)
print(Y)
```

splitting the data into training and testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,random_state = 1,stratify= Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(heart_dataset.shape)
```

Model evaluation

Accuracy testing

```python
from sklearn import svm
model = svm.SVC(kernel= 'linear')
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
X_training_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_training_predict,Y_train)
print("the accuracy is ",training_data_accuracy *100)
```

```python
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict,Y_test)
print("the accuracy is ",test_data_accuracy *100)
```
