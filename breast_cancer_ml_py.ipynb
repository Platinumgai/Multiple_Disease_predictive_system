```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
breast_dataset = pd.read_csv('/content/data (1).csv')
```

```python
breast_dataset['diagnosis'].value_counts()
```

```python
def converter(g):
  if (g == 'M'):
    return 0
  else :
    return 1
breast_dataset['diagnosis'] = breast_dataset['diagnosis'].map(converter)
```

```python
breast_dataset.head()
```

```python
X = breast_dataset.drop(columns = 'diagnosis',axis = 1)
Y = breast_dataset['diagnosis']
```

```python
scalar = StandardScaler()
scalar.fit(X)
std_data = scalar.transform(X)
```

```python
X = std_data
print(X)
print(Y)
```

splitting the data to training and testin g

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model evlauation

accuracy of  the data

```python
from sklearn import svm
model = svm.SVC(kernel ='linear')
model.fit(X_train,Y_train)
```

```python
x_training_predict = model.predict(X_train)
training_data = accuracy_score(x_training_predict,Y_train)
print('The accuraccy of the training data is ', 100 *training_data )
```

```python
X_testing_predict = model.predict(X_test)
test_data = accuracy_score(X_testing_predict,Y_test)
print("the accuracy of the testing data is ",100 *test_data )
```
